```python
#NO1 Remove the following columns from the dataframe:
#new_cases_smoothed, new_deaths_smoothed, new_cases_per_million, total_cases_per_million
```

```python
import pandas as pd
```

```python
df=pd.read_csv('owid-covid-data.csv')
```

```python
df1=df.drop(columns=['new_cases_smoothed','new_deaths_smoothed','new_cases_per_million','total_cases_per_million'])
```

```python
print(df1)
```

```python
# #NO2 Rename the following columns:
# date -> Date
# location -> Country
# Continent -> Continent
# iso_code -> ISO_code
```

```python
df1.rename(columns={'date':'Date','location':'Country','continent':'Continent','iso_code':'ISO_code'},inplace=True)
```

```python
df1
```

```python
#NO3 Using .unique(), create a list of the unique continent names and set it equal to the variable continent_unique
```

```python
uniquecontinent=df1['Continent'].unique()
```

```python
print(uniquecontinent)
```

```python
#NO4  We want to remove these regions so we can just focus on countries
```

```python
df2=df1.dropna(subset=['Continent'])
```

```python
df2
```

```python
#NO5 Create df3 using the groupby and sum commands with the following: reset_index()按照index順序排列
```

```python
df3=df2.groupby(['Date','Country'])[['Date','Country','total_cases','total_deaths','total_vaccinations']].sum().reset_index()
```

```python
df3
```

```python
#NO6 In df3, replace the ‘nan’ values in the Total Cases column with 0 Do the same for the Total Deaths and Total Vaccinations columns
```

```python
df3['total_cases']=df3['total_cases'].fillna(0)
```

```python
df3
```

```python
df3['total_deaths']=df3['total_deaths'].fillna(0)
```

```python
df3
```

```python
df_Mexico=df3[df3['Country'] == 'Mexico']
```

```python
df_Mexico
```

```python
## x = Number of Days Since First Case, y = Number of Cases
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
fig,ax=plt.subplots(figsize=(15,8))
ax.plot(np.arange(0,len(df_Mexico)),df_Mexico['total_cases'],color='red',label='total_cases')
ax.plot(np.arange(0,len(df_Mexico)),df_Mexico['total_deaths'],color='blue',label='total_deaths')
plt.xlabel('Number of Days Since First Case') 
plt.ylabel('Number of Cases')
plt.legend()
plt.title('Mexico')
```

```python
#NO3 Using the code from question 1, create a For Loop that runs the code and outputs a graph for each country in the countries list. 
```

```python

countries =['Mexico','Taiwan','France','Australia']

for i in range(0,len(countries)):
    df=df3[df3['Country']==countries[i]].reset_index()
    plt.scatter(np.arange(0,len(df)),df['total_cases'],label='total_cases',color='black')
    plt.scatter(np.arange(0,len(df)),df['total_deaths'],label='total_deaths',color='yellow')
    plt.title(countries[i])
    plt.xlabel('Number of Days Since First Case')
    plt.ylabel('Number of Cases')
    plt.legend
    plt.show()
```
